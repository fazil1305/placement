```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('placement.csv')
data.head()
```

```python
data.info()
```

```python
data['salary'].isnull().sum()
```

```python
data['salary'].fillna('0',inplace=True)
```

```python
data.info()
```

```python
data.head()
```

```python
data.shape
```

```python
a=pd.get_dummies(data,columns=['gender', 'ssc_b', 'hsc_b', 'hsc_s',
        'degree_t', 'workex',  'specialisation','status'],drop_first=True)
```

```python
a.corr()
```

```python
a.drop(['sl_no','salary','mba_p','ssc_b_Others','hsc_b_Others','specialisation_Mkt&HR','degree_t_Others','degree_t_Sci&Tech','hsc_s_Commerce','hsc_s_Science'],axis=1,inplace=True)
```

```python
a.head()
```

```python
X=a.loc[:,a.columns !='status_Placed']
X.head()
```

```python
y=a.loc[:,'status_Placed']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.3)
```

```python
print(X_train.shape,
X_test.shape,
y_train.shape,
y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfclf=RandomForestClassifier(n_estimators=1000,n_jobs=-1,random_state=42)
```

```python
rfclf.fit(X_train,y_train)
```

```python
accuracy_score(y_test,rfclf.predict(X_test))
```

```python
from sklearn.model_selection import RandomizedSearchCV
rf_clf=RandomForestClassifier(n_estimators=250,random_state=42)
```

```python
param_grid={"max_features":[4,5,6,7,8,9,10],
           "min_samples_split":[2,3,10]}
```

```python
rand_search=RandomizedSearchCV(rf_clf,param_grid,cv=5,n_jobs=-1,scoring="accuracy")
rand_search.fit(X_train,y_train)
```

```python
c=rand_search.best_estimator_
```

```python
rand_search.best_params_
```

```python
accuracy_score(y_test,c.predict(X_test))
```

```python
import xgboost as xgb
xgb_clf=xgb.XGBClassifier(max_depth=5,n_estimators=10000,learning_rate=0.3,n_jobs=-1)
xgb_clf.fit(X_train,y_train)
```
